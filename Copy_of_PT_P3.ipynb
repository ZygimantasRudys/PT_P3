### Step 4. Which continent drinks more beer on average?

```python
# Group data by continent and calculate the mean beer servings
beer_by_continent = drinks.groupby('continent')['beer_servings'].mean()

# Find the continent with the highest average beer servings
continent_with_most_beer = beer_by_continent.idxmax()

# Print the result
print(f"The continent that drinks more beer on average is: {continent_with_most_beer}")
```

### Step 5. For each continent print the statistics for wine consumption.

```python
# Group data by continent and describe wine servings
wine_stats_by_continent = drinks.groupby('continent')['wine_servings'].describe()

# Print the statistics for wine consumption for each continent
print(wine_stats_by_continent)
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
# Group data by continent and calculate mean alcohol consumption
mean_alcohol_by_continent = drinks.groupby('continent').mean()

# Print the mean alcohol consumption per continent for every column
print(mean_alcohol_by_continent)
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
# Group data by continent and calculate median alcohol consumption
median_alcohol_by_continent = drinks.groupby('continent').median()

# Print the median alcohol consumption per continent for every column
print(median_alcohol_by_continent)
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
# Aggregate data to calculate mean, min, and max values for spirit consumption
spirit_stats = drinks.groupby('continent')['spirit_servings'].agg(['mean', 'min', 'max'])

# Print the DataFrame containing mean, min, and max values for spirit consumption
print(spirit_stats)
```

### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
# Importing necessary libraries
import pandas as pd

# Importing the dataset
data_url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data"
data = pd.read_csv(data_url, sep='\s+', parse_dates=[[0, 1, 2]])

# Renaming the combined date column
data.rename(columns={'Yr_Mo_Dy': 'Date'}, inplace=True)

# Setting the Date column as the index
data.set_index('Date', inplace=True)
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def fix_year(date):
    if date.year >= 2061:
        return date.replace(year=date.year - 100)
    else:
        return date

# Apply the fix_year function to the index
data.index = data.index.map(fix_year)
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data.index = pd.to_datetime(data.index)
print(data.index.dtype)
```

### Step 6. Compute how many values are missing for each location over the entire record.  
#### They should be ignored in all calculations below.

```python
missing_values_per_location = data.isnull().sum()
print(missing_values_per_location)
```

### Step 7. Compute how many non-missing values there are in total.

```python
total_non_missing_values = data.notnull().sum().sum()
print(total_non_missing_values)
```

### Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
#### A single number for the entire dataset.

```python
mean_windspeed = data.mean().mean()
print(mean_windspeed)
```

### Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days

#### A different set of numbers for each location.

```python
loc_stats = pd.DataFrame({
    'min': data.min(),
    'max': data.max(),
    'mean': data.mean(),
    'std': data.std()
})

print(loc_stats)
```

### Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

#### A different set of numbers for each day.

```python
day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})

print(day_stats)
```

### Step 11. Find the average windspeed in January for each location.  
#### Treat January 1961 and January 1962 both as January.

```python
# Select rows corresponding to January
january_data = data[data.index.month == 1]

# Calculate the mean windspeed for each location in January
average_windspeed_january = january_data.mean()

print("Average windspeed in January for each location:")
print(average_windspeed_january)
```

### Step 12. Downsample the record to a yearly frequency for each location.

```python
# Downsample the record to a yearly frequency for each location
yearly_mean_windspeed = data.resample('Y').mean()

print("Yearly mean windspeed for each location:")
print(yearly_mean_windspeed)
```

### Step 13. Downsample the record to a monthly frequency for each location.

```python
# Downsample the record to a monthly frequency for each location
monthly_mean_windspeed = data.resample('M').mean()

print("Monthly mean windspeed for each location:")
print(monthly_mean_windspeed)
```

### Step 14. Downsample the record to a weekly frequency for each location.

```python
weekly_data = data.resample('W').mean()

print(weekly_data.head())  # Print the first 5 rows of the weekly data
```

### Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
weekly_data = data.resample('W').mean()

# Calculate the min, max, mean and standard deviation of the windspeeds across all locations for each week
weekly_data['min'] = weekly_data.min(axis=1)
weekly_data['max'] = weekly_data.max(axis=1)
weekly_data['mean'] = weekly_data.mean(axis=1)
weekly_data['std'] = weekly_data.std(axis=1)

# Print the first 52 weeks of the data
print(weekly_data.head(52))
```
